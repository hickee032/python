# 이상치 처리

    특정 지정된 그룹에 분류되지 못하는 값으로, 정상군의 상한과 하한의 범위를 벗어난 자료

사분위수

    전체데이터를 4등분 경계면을 3등분

경계면

    Q1 ~ Q3 : 0.25 - 0.50 - 0.75

IQR 

    Q3 - Q1
    사분위수의 상위 75% 지점의 값과 하위 25% 지점의 값 차이

제거할 이상치 

    (Q3 + IQR * 1.5 보다 큰값) | (Q1 - IQR * 1.5보다 작은값)

    최대 이상치의 경계 : (Q3 + IQR * 1.5 보다 큰값)
    최소 이상치의 경계 : (Q1 - IQR * 1.5보다 작은값)

    -> 곱셈이 먼저 이루어져야 한다

1.5를 곱하는 이유

image.png



```python
# 시리즈
Q1 = df['PBR'].quantile(q=0.25)
print('Q1 : ', Q1)

Q3 = df['PBR'].quantile(q=0.75)
print('Q3 : ', Q3)

IQR = Q3 - Q1
print('IQR : ', IQR)
```

```python
IQR_weight = (Q3-Q1) * 1.5
th_min = Q1 - IQR_weight
th_max = Q3 + IQR_weight

# 이상치 출력
df[(df['PBR'] < th_min) | (df['PBR'] > th_max)]
```

```python
# 이상치 인덱스 추출 
idx = df[(df['PBR'] < th_min) | (df['PBR'] > th_max)].index
print(idx)
print(type(idx))
```

```python
# 이상치 삭제
df.drop(index = idx, inplace=True)
```

```python
def boxplot_scatter(col):
  for i ,d in enumerate(col):
    y = col[d]
    x = np.random.normal(i+1,0.04,len(y))
    plt.scatter(x,y)

df[['PBR']].boxplot()
boxplot_scatter(df[['PBR']])
```
